# Performing a function to extract the exact genre of the movie

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

# Fetching the first three actors from our dataset

```python
def convert3(obj):
    L = []
    counter=0
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

# Fetching the Director from our movie dataset

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head(1)
```

```python
#splitting each word in overview column to get the tags we need
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

# Removing the stop words and spaces

```python
movies['genres'] = movies['genres'].apply(lambda x :[i.replace(" ","")for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x :[i.replace(" ","")for i in x])
movies['cast'] = movies['cast'].apply(lambda x :[i.replace(" ","")for i in x])
movies['crew'] = movies['crew'].apply(lambda x :[i.replace(" ","")for i in x])
```

# Merging all the columns except movie_id and title into a new column named 'tags'

```python
movies['tags'] = movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
#making a new dataframe to merge the date we need

new_df=movies[['movie_id','title','tags']]
```

```python
new_df
```

```python
#converting the data in string format again

new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

# Converting the tags column into lowercase

```python
new_df['tags']= new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

# Performing stemming on our dataset

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
#performing stemming on our dataset in tags column
new_df['tags'].apply(stem)
```

```python
stem('believe')
```

# Vectorizing our dataset using 'bag of words'

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
#converting our data frame into vectors
vectors = cv.fit_transform(new_df['tags']).toarray()
```

# Using cosine distancing

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
#sorting the indexes to get the most similar recommendations first

#used lambda to sort the indexes except the first index

sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:8]
```

# Shaping our final dataset to recommend the most suitable and related results

```python
def recommend(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:8]
    #fetching the titles instead of movie ids
    for i in movies_list:
        print(new_df.iloc[i[0]].title)    
```

```python
#fetching the ids of the same movies

recommend('Batman')

```
